```python
import matplotlib
import matplotlib.pyplot as plt
import os
from PIL import Image  # PIL is the Python Imaging Library, used for opening image files
import kagglehub
import random
import pandas as pd
import panel as pn
import numpy as np
import jupyter_bokeh

# Initialize Panel extension
pn.extension()

# Download latest version
path = kagglehub.dataset_download("ronanpickell/b200c-lego-classification-dataset")

print("Path to dataset files:", path)

BASE_PATH = path  # Update this path

# Find all folders (including nested) within BASE_PATH
available_folders = []
for root, dirs, files in os.walk(BASE_PATH):
    for dir_name in dirs:
        available_folders.append(os.path.relpath(os.path.join(root, dir_name), BASE_PATH))

# Dropdown widget for selecting a folder
folder_selector = pn.widgets.Select(name="Select Folder", options=available_folders)

# Dropdown widget for selecting a file (populated based on the selected folder)
file_selector = pn.widgets.Select(name="Select File", options=[])

# Panel widget to display the selected image
image_display = pn.pane.PNG(width=300)

# Function to update file dropdown based on selected folder
def update_file_options(event):
    selected_folder = folder_selector.value
    folder_path = os.path.join(BASE_PATH, selected_folder)
    # List all images in the selected folder
    image_files = [file for file in os.listdir(folder_path) if file.endswith(('.jpg', '.png'))]
    file_selector.options = image_files

# Attach the update_file_options function to folder_selector
folder_selector.param.watch(update_file_options, 'value')

# Function to display the selected image
def display_selected_image(event):
    selected_folder = folder_selector.value
    selected_file = file_selector.value
    if selected_folder and selected_file:
        image_path = os.path.join(BASE_PATH, selected_folder, selected_file)
        image = Image.open(image_path)
        image_display.object = image

# Attach the display_selected_image function to file_selector
file_selector.param.watch(display_selected_image, 'value')

# Layout
app = pn.Column(
    pn.pane.Markdown("# LEGO Image Viewer"),
    folder_selector,
    file_selector,
    image_display
)

# Update file options initially for the first folder in the list
update_file_options(None)

# Show the app
app.servable()
```
